### [Matplotlib](https://matplotlib.org/)

Visualization in Data Science:
1. Important step in Data Exploration/Analysis - EDA
2. It helps developing an intuition for relationships in data.
3. Precusor to higher level data analysis using ML/DL Techniques.
4. Present/Display Business Insights or Finding to end-user.

**Matplotlib is a Python 2D plotting library which produces publication quality figures in a variety of hardcopy formats and interactive environments across platforms.**

Matplotlib package contains:
1. Pylab Module - MatLab-like graphs and visualization
2. matplotlib.pyplot Module - higher level module which works on abstraction and deals with the data mainly.
3. Object level APIs - core bulding block, axis, titles, plot etc.

It has seamless integration with other packages in PyData Stack - Numpy, Pandas, Scikit-learn etc..

Topics:
1. Anatomy of a Figure
2. Plots, titles, lines, markers, watermarkers
3. texts and Annotations
4. Box-plots and violin Plots
5. Histograms and Pie-charts
6. auto-correlation

### Anatomy of a Figure

**Three Top-Level Components**
1. Pylab -- Collection of command style functions that make matplotlib work like MATLAB.
2. Frontend -- set of classes that abstract actual plotting mechanics and details using a high-level user API.
3. Backend -- Renderers, that transform the frontend representation to hardcopy or a display device.

**Hierarchy**
1. Figure - Top-level Container class, it is divided into Grid as well.
2. Multiple Axes 
3. Multiple-Axis
4. Minor/Major Ticks, Title, Labels, markers, plots, legends

**Modes**
1. Non-Interactive
2. Interactive

```python
x = np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x))

plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('sin(x)', fontsize=15, color='green');
```

#### tick parameters

```python
x = np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x))

plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('sin(x)', fontsize=15, color='green');

plt.tick_params(axis='y',
               color='red',
               labelcolor='blue',
               labelsize='xx-large')

plt.tick_params(axis='x',
               bottom=False,
               labelbottom=False)
```

#### Legend

```python
x = np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x), label='sine curve')

plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('sin(x)', fontsize=15, color='green');

plt.tick_params(axis='y',
               color='red',
               labelcolor='blue',
               labelsize='xx-large')

plt.tick_params(axis='x',
               bottom=False,
               labelbottom=False)

plt.legend()
```

#### title

```python
x = np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x), label='sine curve')

plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('sin(x)', fontsize=15, color='green')
plt.title('Learning basic plots')

plt.tick_params(axis='y',
               color='red',
               labelcolor='blue',
               labelsize='xx-large')

plt.tick_params(axis='x',
               bottom=False,
               labelbottom=False)

plt.legend()
```

#### Set limit on x values

```python
x = np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x), label='sine curve')

plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('sin(x)', fontsize=15, color='green')
plt.title('Learning basic plots')
plt.xlim(1,5)

plt.tick_params(axis='y',
               color='red',
               labelcolor='blue',
               labelsize='xx-large')

plt.tick_params(axis='x',
               color='black',
               labelcolor='blue',
               labelsize='xx-large')

plt.legend()
```

### set limit on y values

```python
x = np.linspace(start=0,stop=10,num=50)
plt.plot(x,np.sin(x), label='sine curve')

plt.xlabel('x', fontsize=15, color='green')
plt.ylabel('sin(x)', fontsize=15, color='green')
plt.title('Learning basic plots')
plt.xlim(1,5)
plt.ylim(-1,.5)

plt.tick_params(axis='y',
               color='red',
               labelcolor='blue',
               labelsize='xx-large')

plt.tick_params(axis='x',
               color='black',
               labelcolor='blue',
               labelsize='xx-large')

plt.legend()
```

```python
type(ax1)
```

```python
x = np.linspace(start=0,stop=10,num=50)

fig = plt.figure(figsize=(8,8))
ax1 = fig.add_subplot(221) #2*2 Grid, 1 cell
ax1.plot([2,4,6,8],[4,8,12,16])

ax2 = fig.add_subplot(222)
ax2.plot(x,np.sin(x))

ax3 = fig.add_subplot(223)
ax3.plot(x,np.cos(x))

ax4 = fig.add_subplot(224)
ax4.plot(x,np.exp(x))
```

```python
x = np.linspace(start=0,stop=10,num=50)

ax1 = plt.subplot2grid((2,3),(0,0))
ax1.plot([2,4,6,8],[4,8,12,16])

ax2 = plt.subplot2grid((2,3),(0,1))
ax2.plot(x,np.sin(x))

ax3 = plt.subplot2grid((2,3),(0,2), rowspan=2)
ax3.plot(x,np.cos(x))

ax4 = plt.subplot2grid((2,3),(1,0), colspan=2)
ax4.plot(x,np.exp(x))
plt.tight_layout()
```

### watermarks

```python
fig, ax = plt.subplots()

ax.plot([1,2,3,4],[2,4,6,8])

ax.text(1,4,'confidential',
       fontsize=30,
       color='red',
       ha='left',
       va='bottom',
       alpha=0.5)
```

```python
x = np.linspace(start=0,stop=10,num=50)

ax1 = plt.subplot2grid((2,3),(0,0))
ax1.plot([2,4,6,8],[4,8,12,16])

ax2 = plt.subplot2grid((2,3),(0,1))
ax2.plot(x,np.sin(x))

ax3 = plt.subplot2grid((2,3),(0,2), rowspan=2)
ax3.plot(x,np.cos(x))

ax4 = plt.subplot2grid((2,3),(1,0), colspan=2)
ax4.plot(x,np.exp(x))

ax4.text(2,1000,'confidential',
       fontsize=10,
       color='red',
       ha='left',
       va='bottom',
       alpha=0.5)
plt.tight_layout()
```
